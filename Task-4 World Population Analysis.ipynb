# Which Countries have the largest population?

```python
wp_df.set_index("Country/Territory", inplace=True)
plt.title('Countries with Largest Population')
plt.xlabel("Country/Territory")
plt.ylabel("No.Population")
wp_df['2020 Population'].nlargest().plot(kind = "bar" ,color="green")
```

# Which country with the biggest Density? India or China?

```python
india = (wp_df.loc["India"])["Density (per km²)"]
china = (wp_df.loc["China"])["Density (per km²)"]
print("The Density in India = " ,india)
print("The Density in China = " ,china)
```

```python
area1 = (wp_df.loc["India"])["Area (km²)"]
area2 = (wp_df.loc["China"])["Area (km²)"]
countries = [india,china]
area = [area1,area2]
plt.xlabel("Density")
plt.ylabel("Area")
plt.plot(countries,area, "-g" , linewidth=5 ,linestyle="solid")
plt.show()
```

# What do you predict for the Density in the next four years?

```python
population_2030 = wp_df["2020 Population"] * (wp_df["Growth Rate"] ** 10)
population_2040 = wp_df["2020 Population"] * (wp_df["Growth Rate"] ** 20)
population_2050 = wp_df["2020 Population"] * (wp_df["Growth Rate"] ** 30)
population_2060 = wp_df["2020 Population"] * (wp_df["Growth Rate"] ** 40)
print("The Density in 2030 is " , population_2030)
print("The Density in 2040 is " , population_2040)
print("The Density in 2050 is " , population_2050)
print("The Density in 2060 is " , population_2060)
```

```python
wp_df["population_2030"] = population_2030
wp_df["population_2040"] = population_2040
wp_df["population_2050"] = population_2050
wp_df["population_2060"] = population_2060
```

```python
wp_df
```

# let's see what happen to Population in India and China

```python
india_population_2030 = (wp_df.loc["India"])["population_2030"]
China_population_2030 = (wp_df.loc["China"])["population_2030"]
india_population_2040 = (wp_df.loc["India"])["population_2040"]
China_population_2040 = (wp_df.loc["China"])["population_2040"]
india_population_2050 = (wp_df.loc["India"])["population_2050"]
China_population_2050 = (wp_df.loc["China"])["population_2050"]
india_population_2060 = (wp_df.loc["India"])["population_2060"]
China_population_2060 = (wp_df.loc["China"])["population_2060"]
```

```python
india__population = [india_population_2030,india_population_2040,india_population_2050,india_population_2060]
years = [2030,2040,2050,2060]
china__population = [China_population_2030,China_population_2040,China_population_2050,China_population_2060]
```

```python
plt.style.use("dark_background")
plt.title("What Happen for Population in India at the next 4 years")
plt.plot(years,india__population ,color="g" ,linewidth=5 ,linestyle="solid" ,label="India_population")
plt.plot(years,china__population ,color="orange" ,linewidth=5 ,linestyle="--" ,label="China_population")
plt.xlabel("years")
plt.ylabel("population")
plt.legend()
plt.show()
```
